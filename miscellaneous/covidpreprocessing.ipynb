```python
import os
import numpy as np
import pandas as pd
import cv2
import PIL
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
train_path = '../input/covidx-cxr2/train/'  #directory path
test_path = '../input/covidx-cxr2/test/'
```

```python
def read_record(record):
    x=tf.strings.split(record,sep=' ')
    return x[1], x[2]

def train_image(image_file, label):
    #image_path = train_path + image_file
    image_path = tf.strings.join([train_path,image_file])
    image = (tf.io.decode_image(tf.io.read_file(image_path), channels=3,expand_animations=False))/255
    image = tf.image.resize(image, [128,128])
    label = 0 if label=='negative' else 1 
    label = tf.one_hot(label, depth=2)
    return image, label

def test_image(image_file, label):
    #image_path = train_path + image_file
    image_path = tf.strings.join([test_path,image_file])
    image = (tf.io.decode_image(tf.io.read_file(image_path), channels=3,expand_animations=False))/255
    image = tf.image.resize(image, [128,128])
    label = 0 if label=='negative' else 1 
    label = tf.one_hot(label, depth=2)
    return image, label
```

```python
dataset = tf.data.TextLineDataset(["../input/covidx-cxr2/train.txt"])
dataset = dataset.map(read_record).map(train_image).batch(32)
```
